```python
# FACE EMOTION DETECTION MACHINE LEARNING PROJECT
#JoshuaMaeba codes
# import the required packages
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
# Importing our image
img = cv2.imread("/content/bsc.statistics.PNG")
```

```python
# Showing the image in BGR format
plt.imshow(img)
```

```python
# Show the image in  it's original color
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
predictions = DeepFace.analyze(img)
predictions
```

```python
#Check Dominant emotion
predictions[ 'dominant_emotion']
```

```python
# Drawing a rectangle accross the face
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_default.xml")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Print facecascade empty
faces = faceCascade.detectMultiScale(gray, 1.1,4)
# Draw rectangle round the face
for (x,y,w,h) in faces:
  cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0),2)

```

```python
# Show the image with rectangle around the face
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Insert the dominant emotion on the rectangle
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,
            predictions['dominant_emotion'],
            (50,50),
            font,1,
            (0,0,255),
            2,
            );
```

```python
# Show the image with the dominant emotion tag
# Show the image in  it's original color
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
